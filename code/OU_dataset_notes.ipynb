### Data

* student demographic data
* vle data (potentially dynamic - timeseries)

#### Models

* several predictive models
* combined to classify students at risk


### Purpose

* early identification of students who are at risk of 'failing', 'withdrawing' or 'not attaining' (potential other categories - successes, completions, progressions)
* ultimately - improving retention by supporting students (business)

#### Opportunities
* recommendation of interventions (activities, resources, support)
* individual learning plans
* programme design
